```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt



#Importing the dataset
dataset = pd.read_csv('50_Startups.csv')

#print(dataset)


RD = dataset['R&D Spend'].values.reshape(-1,1)
ADMIN = dataset['Administration'].values.reshape(-1,1)
MARKET = dataset['Marketing Spend'].values.reshape(-1,1)
y = dataset['Profit'].values.reshape(-1,1)




RD_train, RD_test, y_train, y_test = train_test_split(RD, y, test_size=0.2, random_state=1)

RDregressor = LinearRegression()  
RDregressor.fit(RD_train, y_train) #training the algorithm

RD_prediction = RDregressor.predict(RD_test)
```

```python
ADM_train, ADM_test, y_train, y_test = train_test_split(ADMIN, y, test_size=0.2, random_state=1)

ADMregressor = LinearRegression()  
ADMregressor.fit(ADM_train, y_train) #training the algorithm

ADM_prediction = ADMregressor.predict(ADM_test)
```

```python
MAR_train, MAR_test, y_train, y_test = train_test_split(MARKET, y, test_size=0.2, random_state=1)

MARregressor = LinearRegression()  
MARregressor.fit(MAR_train, y_train) #training the algorithm

MAR_prediction = MARregressor.predict(MAR_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'RD Predicted': RD_prediction.flatten(), 'ADM Predicted': ADM_prediction.flatten(), 'MAR Predicted': MAR_prediction.flatten()})
df
```
