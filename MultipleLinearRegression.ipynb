Importer alt og forbered train og test data


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

linear_regression1 = LinearRegression()
linear_regression2 = LinearRegression()
linear_regression3 = LinearRegression()
linear_regression4 = LinearRegression()

#Importing the dataset
dataset = pd.read_csv('50_Startups.csv')

#turn states into numbers
dataset['State']=dataset['State'].apply(lambda x:float(x.replace("New York","1").replace("Florida", "2").replace("California", "3")))

y = dataset['Profit'].values.reshape(-1,1)
X1 = dataset[['R&D Spend', 'Administration', 'Marketing Spend']].values.reshape(-1,3)
X2 = dataset[['R&D Spend', 'Marketing Spend']].values.reshape(-1,2)
X3 = dataset[['R&D Spend']].values.reshape(-1,1)
X4 = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State']].values.reshape(-1,4)

```

```python
linear_regression1.fit(X1,y)
linear_regression2.fit(X2,y)
linear_regression3.fit(X3,y)
linear_regression4.fit(X4,y)
y_prediction1 = linear_regression1.predict(X1)
y_prediction2 = linear_regression2.predict(X2)
y_prediction3 = linear_regression3.predict(X3)
y_prediction4 = linear_regression4.predict(X4)


```

Here comes the results

```python
print("R2 score with three parameters: "+str(r2_score(y, y_prediction1)))
print("R2 score with only R&D and marketing: "+str(r2_score(y, y_prediction2)))
print("R2 score with R&D only: "+str(r2_score(y, y_prediction3)))
print("R2 score with three parameters plus state: "+str(r2_score(y, y_prediction4)))

print(" ")

# Koefficienten (altså modellen)
print('Coefficients for three parameters: ', linear_regression1.coef_)
print('Coefficients for only R&D and marketing: ', linear_regression2.coef_)
print('Coefficients for R&D only: ', linear_regression3.coef_)
print('Coefficients for three parameters plus state: ', linear_regression4.coef_)

print(" ")



df = pd.DataFrame({'Actual': y.flatten(), 'Prediction with admin': y_prediction1.flatten(), 'Prediction without admin': y_prediction2.flatten(), 'Prediction with R&D alone:': y_prediction3.flatten(), 'Prediction with everything': y_prediction4.flatten()})
df
```
